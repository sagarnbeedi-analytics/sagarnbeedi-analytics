### CREATING SERIES IN PANDAS
when we create the series using pandas we get the index values and data type of the series as output and we use series function of pandas to create the series data struture

creating array using the list

```python
s1=pd.Series([10,20,30,40,50])
print(s1)
```

### PROPERTIES OF THE SERIES 
There are several properties of the series which can be checked such as data type values name and index of the series as given below 

initially the name of series is none and we can assign the desired name to series and when we access the series after that it will be denoted by that name 

```python
print(s1.dtype)
print(s1.values)
print(s1.index)
print(s1.name)
s1.name="calaries"
print(s1.name)
print(s1)
```

### INDEXING IN SERIES DATA STRUTURE IN PANDAS

indexing in series is same like the python lists or numpy arrays. In series you can see the values at certain index based on range you give the start index and end index where end index is not including 

```python
s2=pd.Series([1,2,3,4,5])
print(s2[0])
print(s2[0:3])
print(s2[1:3])
```

indexing using the iloc function which does indexing based on location of the value so iloc function does the location based indexing which means it will use indexes 0,1,2,3,4 etc

```python
print(s1.iloc[2])
print(s2.iloc[2])
```

iloc function is used when you want multiple values which are on different numerical indexes as given below 

```python
print(s2.iloc[[2,3]])
```

### CUSTOM INDEXING IN SERIES 

custom indexing is done in series when you dont want palin number as indext of the values and want to put some other data as index so that the series makes more sense.

we create the list and assign the list to series index to perform the custom indexing

custom indexing is done to represent the c=indexes in the more meaningful way so that the coulmn gives important info

```python
print(s1)
idx=['apple','mango','banana','orange','pineapple']
s1.index=idx
print(s1)
```

finding values of the series using custom indexes 

```python
print(s1['apple'])
```

iloc function is used to get the values at the numerical indexes but not for string indexes. For string indexing or custom indexing which is not number you can using loc function as given below 

remember that in label based indexing both start and end index is inclued where as in location based indexing end value is not incliuded 

```python
print(s1.loc[['apple','mango']])
print(s1.loc[['apple','mango','pineapple']])
```

### LOGICAL OPERATORS FOR PANDAS SERIES
logical operators are logical and logical or and logical not which can be used on series as well. We use & for and operator and | for or perator

for logical and we must have both condition true to get true result where as for or operator we need any one condition to be true to the output based on that true condition 

```python
mask=s3[(s3>0.5)&(s3<2)]
print(mask)
```

```python
mask=s3[(s3>0.5) | (s3<2)]
print(mask)
```

```python
print(s3[s3!=1])
```

### DATAFRAME DATA STRUCTURE IN PANDAS 

we can convert the dictionary into data frames so that they can be easily be represented properly in the form of table. A dataframe is a two dimensional data struture in pandas where rows and columns both are accepted 

```python
data={
    'names':['sagar','rahul','akshay','reshma','divya','ramesh','suresh'],
    'Age':[34,31,33,28,25,54,45],
    'department':['IT','HR','Finance','Admin','marketing','sales','managemnt']
}
empdata=pd.DataFrame(data)
print(empdata)
```
