FORWARD AND BACKWORD FILLS FOR MISSING VALUES

if we use backward ifll the value after the missing value is filled in the missing value 

FORWARD FILL:

```python
empdata['Salary'].fillna(method='ffill')
empdata['Salary'].fillna(method='bfill')
```

### DEALING WITH DUPLICATE DATA IN DATAFRAME

CHECKING FOR DUPLICATE VALUES:

The sample is only categorized as duplicate if and only if all the features of that sample are same with another sample 

There is a keep parameter for duplicated function which can take first and last as argument which means it will not mark first element as duplicate and when we give last it will not mark last element as duplicate 

```python
dup=empdata[empdata.duplicated(keep='first')]
dup=empdata[empdata.duplicated(keep='last')]
dup
```
